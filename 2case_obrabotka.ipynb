```python
import seaborn as sns
import matplotlib.pyplot as plt

df = df = pd.read_excel('distribution.xlsx')
plt.figure(figsize=(12, 8))
pivot_table = df.pivot_table(index='driver_id', 
                            columns='order_id',
                            values='distance',
                            aggfunc='first')

sns.heatmap(pivot_table, cmap='YlOrRd', cbar_kws={'label': 'Distance (hexes)'})
plt.title('Распределение заказов между водителями')
plt.xlabel('Order ID')
plt.ylabel('Driver ID')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='distance', y='start_price', hue='driver_rating', 
                palette='coolwarm', size='driver_rating', sizes=(20, 200))
plt.title('Соотношение расстояния и цены заказа')
plt.xlabel('Расстояние (гексы)')
plt.ylabel('Цена заказа')
plt.grid(True)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df['driver_id'].value_counts().value_counts().sort_index().plot(kind='bar')
plt.title('Распределение количества заказов на водителя')
plt.xlabel('Количество заказов')
plt.ylabel('Число водителей')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
driver_stats = df.groupby('driver_id').agg({
    'distance': 'mean',
    'start_price': 'sum',
    'driver_rating': 'first'
}).sort_values('start_price', ascending=False)

plt.figure(figsize=(12, 6))
sns.scatterplot(data=driver_stats, x='distance', y='start_price', hue='driver_rating',
                palette='viridis', size='driver_rating', sizes=(50, 200))
plt.title('Производительность водителей')
plt.xlabel('Среднее расстояние (гексы)')
plt.ylabel('Общий заработок')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df['order_type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Распределение по типам заказов')
plt.ylabel('')
plt.show()
```

```python
top_drivers = df.groupby('driver_id').agg({
    'start_price': 'sum',
    'driver_rating': 'first'
}).nlargest(10, 'start_price')

plt.figure(figsize=(12, 6))
sns.barplot(data=top_drivers.reset_index(), x='driver_id', y='start_price', hue='driver_rating', dodge=False)
plt.title('Топ-10 водителей по заработку')
plt.xlabel('ID водителя')
plt.ylabel('Общий заработок')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
